2. ARPU

```python
# tng doanh thu
revenue_mar = receipts_history[(receipts_history['date'].dt.month==3)& (receipts_history['date'].dt.year==2023)]['receipt_amount'].sum()
```

```python
# s lng khch hng
customer_mar = receipts_history[(receipts_history['date'].dt.month==3)&(receipts_history['date'].dt.year==2023)]
customer_number = customer_mar['customer_id'].count()
print(customer_number)
```

```python
ARPU = revenue_mar/customer_number
print(ARPU.round(2))
```

3.COGS:

3.1 server and software licences

```python
#server
server= monthly_expenses[monthly_expenses["month"]=="2023-03-01"]
server_cost = server[server['item'].isin(['AWS Hosting','Google Cloud Storage','Atlassian Jira'])]['amount'].sum()

#software: slack, zoom (60%)
software = monthly_expenses[monthly_expenses["month"]=="2023-03-01"]
software_cost = software[software['item'].isin(['Slack','Zoom'])]['amount'].sum()*0.6

```

3.2 production team's salary

```python
last_month_payroll_salary=last_month_payroll[last_month_payroll['department']=='Engineering']['paid'].sum()
```

```python
COGS = server_cost + software_cost + last_month_payroll_salary
print(COGS)
```

4. Gross Margin:

```python

gross_margin = (revenue_mar - COGS)/revenue_mar*100
print(gross_margin.round(2))
```

5. LTV

```python
#Tnh Customerlifespan:
customer_lifespan['lifespan']=(customer_lifespan['churn_date']-customer_lifespan['start_date']).dt.days
customer_life_span=customer_lifespan['lifespan'].sum()/customer_lifespan['lifespan'].count()
LTV = ARPU * (customer_life_span/30) * gross_margin/100
print('LTV is:',LTV.round(2))
```

6. LTV/CAC

```python
LTV_per_CAC = LTV/CAC
print(LTV_per_CAC.round(2))
```
